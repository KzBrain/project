```r
set.seed(42)
trainIndex <- createDataPartition(soybean_data$class,p =.7,list = FALSE,times = 1)
```

soybean_data$class: Целевая переменная, по которой будет выполнено  разделение на тестовые и тренировычные данные при помощи индексов.70% из них уйдет на обучающую выборку.

```r
trainIndex
```

```r
train_data <- soybean_data[trainIndex,]
test_data <- soybean_data[-trainIndex,]
```

```r
head(train_data)
```

```r
install.packages("randomForest")
```

```r
library(randomForest)
```

```r
model <- train(class ~ .,data = train_data,method = "rf",trControl = trainControl(method = "cv",number = 10))
```

Здесь попробовали использовать модель Случайный лес.

```r
print(model)
```

 На основе этих результатов выбран оптимальный параметр mtry = 112

```r
importance <- varImp(model, scale = FALSE)
plot(importance)
```

Мы хотели узнать важность каждого признака.

```r
predictions <- predict(model,test_data)
```

```r
confusions <- confusionMatrix(predictions,test_data$class)
print(confusions)
```

Модель показывает высокую точность (97.8%) и значительную согласованность (каппа = 0.9104).

Чувствительность и специфичность высоки для всех классов, особенно для классов 2 и 4.

Прогнозы модели для классов 2 и 4 практически безошибочны, с точностью и отзывом равными 100%.

Класс 1 имеет меньшую чувствительность (50%), что указывает на трудности модели в правильной классификации всех примеров этого класса, несмотря на высокую специфичность и точность.


```r
library(pROC)
```

```r
proc_obj <- roc(test_data$class,as.numeric(predictions))
```

```r
plot.roc(proc_obj,col = "blue", main = "ROC-кривая для модели Random Forest")
```

 ROC-кривая, которая очень близка к верхнему левому углу, указывает на высокую производительность модели. В данном случае кривая достигает верхнего левого угла, что свидетельствует о высоких значениях как чувствительности, так и специфичности.

```r
result <- model$results
```

```r
ggplot(result,aes(x = mtry,y = Accuracy))+
  geom_line()+
  geom_point()+
  labs(title = "Accuracy в зависимости от количества деревьев", x = "Количество деревьев", y = "Accuracy")
```

На графике видно, что увеличение количества деревьев приводит к почти линейному росту точности. Это указывает на то, что добавление большего количества деревьев продолжает улучшать производительность модели.

```r
cv_model <- train(class ~.,data = train_data,method = "rf",trControl = trainControl(method = "cv",number = 10))
```

```r
plot(test_data$precip, test_data$date, col = ifelse(predictions == test_data$class, "blue", "red"),
     pch = 19, main = "Scatter Plot истинных значений и предсказаний",
     xlab = "precip", ylab = "date")
     legend("topright", legend = c("Правильные", "Неправильные"),
       col = c("black", "red"), pch = 19, bty = "n")
```

Большинство точек на графике синего цвета, что указывает на высокую точность модели.Небольшое количество красных точек указывает на ошибки модели, что согласуется с ранее представленной точностью (около 97.8%).
