```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('/content/sample_data/gathered_2019-2023.xlsx')
```

```python
df.head()
```

```python
dt = pd.read_excel('/content/sample_data/gathered_date.xlsx')
```

```python
dt.head()
```

```python
df_2019 = df[['Death_rate_in_hospital_2019']]
```

```python
df_2019.head()
```

```python
dt_2019 = dt[['Date_start_2019','Date_event_2019']]
```

```python
dt_2019.head()
```

```python
df_main_2019 = pd.concat([df_2019,dt_2019], axis=1)
```

```python
df_main_2019.head()
```

```python
df_main_2019['Death_rate_in_hospital_2019'] = df_main_2019['Death_rate_in_hospital_2019'].replace({'Выписан': 0, 'Умер': 1})
```

```python
df_main_2019.head()
```

```python
df_main_2019['Death_rate_in_hospital_2019'] = pd.to_numeric(df_main_2019['Death_rate_in_hospital_2019'], errors='coerce')
df_main_2019['Death_rate_in_hospital_2019'].hist()
```

```python
df_main_2019['Date_start_2019'] = pd.to_datetime(df_main_2019['Date_start_2019'], format='%d.%m.%Y')
df_main_2019['Date_event_2019'] = pd.to_datetime(df_main_2019['Date_event_2019'], format='%d.%m.%Y')
```

```python
df_main_2019['time'] = (df_main_2019['Date_event_2019'] - df_main_2019['Date_start_2019']).dt.days / 365.25
```

```python
df_main_2019.head()
```

```python
df_main_2019['Death_rate_in_hospital_2019'].isnull().sum()
```

```python
df_main_2019 = df_main_2019.dropna(subset=['time', 'Death_rate_in_hospital_2019'])
```

```python
!pip install lifelines
```

```python
from lifelines import KaplanMeierFitter
```

```python
kmf = KaplanMeierFitter()
kmf.fit(df_main_2019['time'], event_observed= df_main_2019['Death_rate_in_hospital_2019'])

# Визуализация
import matplotlib.pyplot as plt
kmf.plot_survival_function()
plt.title("Кривая Каплана-Майера")
plt.xlabel("Время")
plt.ylabel("Вероятность выживания")
plt.show()
```

```python
kmf_2019 = kmf.fit(df_main_2019['time'], event_observed= df_main_2019['Death_rate_in_hospital_2019'])
```

#Logrank test

```python
from lifelines.statistics import logrank_test
```

```python
def prepare_km_data(df, year):
    df[f'Date_start_{year}'] = pd.to_datetime(df[f'Date_start_{year}'], format='%d.%m.%Y')
    df[f'Date_event_{year}'] = pd.to_datetime(df[f'Date_event_{year}'], format='%d.%m.%Y')
    df['time'] = (df[f'Date_event_{year}'] - df[f'Date_start_{year}']).dt.days / 365.25
    df = df.dropna(subset=['time', f'Death_rate_in_hospital_{year}'])
    return df['time'], df[f'Death_rate_in_hospital_{year}']
```

```python
time_2019, event_2019 = prepare_km_data(df_main_2019, 2019)
time_2020, event_2020 = prepare_km_data(df_main_2020, 2020)
time_2021, event_2021 = prepare_km_data(df_main_2021, 2021)
time_2023, event_2023 = prepare_km_data(df_main_2023, 2023)
```

```python
results_2019_2020 = logrank_test(time_2019, time_2020, event_observed_A=event_2019, event_observed_B=event_2020)
results_2019_2021 = logrank_test(time_2019, time_2021, event_observed_A=event_2019, event_observed_B=event_2021)
results_2019_2023 = logrank_test(time_2019, time_2023, event_observed_A=event_2019, event_observed_B=event_2023)
results_2020_2021 = logrank_test(time_2020, time_2021, event_observed_A=event_2020, event_observed_B=event_2021)
results_2020_2023 = logrank_test(time_2020, time_2023, event_observed_A=event_2020, event_observed_B=event_2023)
results_2021_2023 = logrank_test(time_2021, time_2023, event_observed_A=event_2021, event_observed_B=event_2023)
```

```python
print("Log-rank test results:")
print(f"2019 vs 2020: p-value = {results_2019_2020.p_value}")
print(f"2019 vs 2021: p-value = {results_2019_2021.p_value}")
print(f"2019 vs 2023: p-value = {results_2019_2023.p_value}")
print(f"2020 vs 2021: p-value = {results_2020_2021.p_value}")
print(f"2020 vs 2023: p-value = {results_2020_2023.p_value}")
print(f"2021 vs 2023: p-value = {results_2021_2023.p_value}")
```
