```python
from scipy.stats import f_oneway
```

```python
subset_data = df_main[['Phenotype', 'FRS-R', 'ECAS']].dropna()
```

```python
# 1. Преобразуем Phenotype в числовые значения (Ordinal Encoding)
subset_data['Phenotype_encoded'] = subset_data['Phenotype'].astype('category').cat.codes
```

```python
# Рассчитаем корреляцию Спирмена между закодированным Phenotype и числовыми переменными
spearman_corr_frs = subset_data[['Phenotype_encoded', 'FRS-R']].corr(method='spearman').iloc[0, 1]
spearman_corr_ecas = subset_data[['Phenotype_encoded', 'ECAS']].corr(method='spearman').iloc[0, 1]
```

```python
spearman_corr_frs
```

```python
spearman_corr_ecas
```

```python
# 2. Проведем ANOVA для числовых переменных между группами Phenotype
anova_frs = f_oneway(
    *[subset_data.loc[subset_data['Phenotype'] == category, 'FRS-R'] for category in subset_data['Phenotype'].unique()]
)
anova_ecas = f_oneway(
    *[subset_data.loc[subset_data['Phenotype'] == category, 'ECAS'] for category in subset_data['Phenotype'].unique()]
)
```

```python
anova_frs
```

```python
anova_ecas
```
